```python
fname2 = path+'LCDM_pks_128_L15_zi30_match_track_z5p000.0' 
unit_base = {'UnitLength_in_cm'         : 3.08568e+21}     ##set the units 
 #            'UnitMass_in_g'            :   1.989e+43,
  #           'UnitVelocity_in_cm_per_s' :      100000}

#bbox_lim = 1e5                                           #kpc should encapsulate all the particles in the data set

#bbox = [[-bbox_lim,bbox_lim],
      #  [-bbox_lim,bbox_lim],
       # [-bbox_lim,bbox_lim]]
 
ds2 = yt.load(fname2,unit_base=unit_base)                    #Read the data set in Gadget format
ds2.index           
ad2= ds2.all_data()
```

```python
# Load the dataset
fname = path+'LCDM_pks_128_L15_zi30_match_track_z0p000.0' 
unit_base = {'UnitLength_in_cm'         : 3.08568e+21}     ##set the units 
 #            'UnitMass_in_g'            :   1.989e+43,
  #           'UnitVelocity_in_cm_per_s' :      100000}

#bbox_lim = 1e5                                           #kpc should encapsulate all the particles in the data set

#bbox = [[-bbox_lim,bbox_lim],
      #  [-bbox_lim,bbox_lim],
       # [-bbox_lim,bbox_lim]]
 
ds = yt.load(fname,unit_base=unit_base)                    #Read the data set in Gadget format
ds.index           
ad= ds.all_data()
```

**Graficamos de la proyección de las partículas sobre el plano xy**


```python
p1 = yt.ParticlePlot(ds2,'particle_position_x', 'particle_position_y', 'particle_mass') #Proyection of particle field
                                                                                         #IC
p1.set_axes_unit('Mpc')  #Set the unit for display on the x and y axes of the image.
p1.set_unit('particle_mass', 'Msun/h')
p1.annotate_title('z=5')
```

```python
p2 = yt.ParticlePlot(ds,'particle_position_x', 'particle_position_y', 'particle_mass') #Proyection of particle field

p2.set_axes_unit('Mpc')  #Set the unit for display on the x and y axes of the image.
p2.set_unit('particle_mass', 'Msun/h')
p2.annotate_title('z=0')
```

```python
sorted(ds.field_list) # show a list of the available fields
```

```python
p = yt.ParticleProjectionPlot(ds,1, 'particle_mass',width=6, depth=20) #Make a projection with a 
                                                                               #slice along (0=x, 1=y, 2=z) 
                                                                              #of the x(y or z) axis 
                                                                               #width in the x and y directions
                                                                                # in code units
p.show()
```
